```python
import pandas as pd 

df = pd.read_csv(r"C:/Users/HP/Downloads/automobile/imports-85.data",header=None) 

df.columns=['symboling', 'normalized_losses', 'make', 'fuel_type', 'aspiration', 'num_of_doors', 'body_style', 'drive_wheels', 'engine_location', 'wheel_base', 'length', 'width', 'height', 'curb_weight', 'engine_type', 'num_of_cylinder', 'engine_size', 'fuel_system', 'bore', 'stroke', 'compression_ratio', 'horsepower', 'peak_rpm', 'city_mpg', 'highway_mpg', 'price']
print(df)

```

```python
continous_values=['normalized_losses', 'wheel_base', 'length', 'width', 'height', 'curb_weight', 'engine_size', 'bore', 'stroke', 'compression_ratio', 'horsepower', 'peak_rpm', 'city_mpg', 'highway_mpg', 'price']
numeric_df=df[continous_values]
print(numeric_df)
```

```python
import numpy as np
numeric_df=numeric_df.replace("?", np.nan)
numeric_df.head(5)
```

```python
numeric_df=numeric_df.astype('float')
numeric_df.isnull().sum()
```

```python
numeric_df=numeric_df.dropna(subset=['price'])
numeric_df.isnull().sum()
```

```python
numeric_df=numeric_df.fillna(numeric_df.mean())
numeric_df.info()
```

https://analystanswers.com/data-normalization-techniques-easy-to-advanced-the-best/



```python
normalized_df=(numeric_df-numeric_df.min())/(numeric_df.max()-numeric_df.mean())
normalized_df['price']=numeric_df['price']
normalized_df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
plt.figure(figsize=(16, 6))
mask = np.triu(np.ones_like(normalized_df.corr(), dtype=np.bool_))
heatmap = sns.heatmap(normalized_df.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Triangle Correlation Heatmap', fontdict={'fontsize':18}, pad=16);
```

Correlation yang jelas meliputi; 
City_MPG and Highway_MPG.
Kita juga dapat melihat dibagian pojok kiri atas yang memiliki korelasi yang sangat jelas itu pada Wheel_base, lenght, width, and curb_weight.

Selanjutnya melihat semua korelasi terhadap harga mobil(Price). 




```python
plt.figure(figsize=(8, 12))
heatmap = sns.heatmap(normalized_df.corr()[['price']].sort_values(by='price', ascending=False), vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title("All Features and their Correlation with 'price'", fontdict={'fontsize':18}, pad=16);
```

Pada korelasi diatas yang memiliki nilai korelasi paling tinggi terhadap harga meliputi;
engine_size, curb_weight, and horsepower. Sedangkan yang tidak ada korelasinya itu city_mpg dan juga highway_mpg.
